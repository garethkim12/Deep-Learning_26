```python
import numpy as np

import torch
```

```python
X = np.array([160, 170, 180, 190])
y = np.array([0, 0, 1, 1])
```

```python
X_mean = np.mean(X)

X_std = np.std(X)

X_norm = (X - X_mean) / X_std

X_norm 
```

```python
X_norm_tensor = torch.tensor(X_norm, dtype= torch.float32)

X_norm_tensor
```

```python
y_tensor = torch.tensor(y, dtype=torch.float32)

y_tensor
```

```python
X_norm_tensor
```

```python
X_norm_tensor = X_norm_tensor.reshape(-1, 1)#2차원으로 바꾸는 것 .reshape() 

X_norm_tensor
```

```python
y_tensor
```

```python
y_tensor = y_tensor.reshape(-1, 1) # 찐 값도 2차원 배열 정리

y_tensor
```

```python
torch.manual_seed(42)
```

```python
linear = torch.nn.Linear(1,1)
```

```python
linear.bias
```

```python
linear.bias
```

```python
criterion = torch.nn.BCELoss()

print("criterion 준비 완료:", criterion)
```

```python
learning_rate = 0.1
epochs = 1000
```

```python
linear.parameters()
```

```python
optimizer = torch.optim.SGD(linear.parameters(), lr = learning_rate)
optimizer
```

```python
list( linear.parameters())
```

```python
for epoch in range(epochs): 
    optimizer.zero_grad()
    
    H = linear(X_norm_tensor)
    
    
    z = torch.sigmoid(H)
    
    mean_cost = criterion(z, y_tensor) #z 값 시그모이드
    
    mean_cost.backward()
    optimizer.step()
    
    if epoch % 100 == 0 or epoch == epochs: # epochs 는 마지막일때
        print(f"epoch = {epoch}, cost = {mean_cost}, a = {linear.weight.item()}, b = {linear.bias.item()}")
        
    if epoch < 3:
        print(f"linear.weight.grad = {linear.weight.grad},"
              f"linear.bias.grad = {linear.bias.grad},"
              )
```
