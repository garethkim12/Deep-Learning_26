```python
# 온도에 따라서 떡볶이가 잘 팔림 (예시: 딥러닝) 
import pandas as pd

df = pd.DataFrame ({
    "temperature":[10, 7, 4, 0],
    "sales":[30, 50, 60, 80]
})

display(df)
```

```python
#기온이 1도가 높아지면 더워지니 안팔리는 추세
a = -4
b = 80
error_sum = 0 # 합 값 초기화

print(f"len(df) = {len(df)}\n")

for i in range(len(df)):
    print(f"i ={i}")
    x = df.loc[i, "temperature"] # 온도를 가져옴 [i] 값
    print(f" x={x}\n")
    pred = a * x + b
    print(f"X={x} 일때 예측값 = {a}*{x}+{b} = {pred}\n") 
    y = df.loc[i, "sales"]
    print(f"찐 y= {y}")
    error = y - pred 
    print(f"오차 = {y}-{pred}={error}")
    error_square = error ** 2 # 값이 틀렸을때 더 많이 차감하려고 제곱 진행
    print(f"error_square={error_square}")
    error_sum += error_square #error_square 의 합 
    print("=" * 30)
    
print(f"error_sum = {error_sum}")
cost = error_sum / len(df) #비용 값은 loss 비용
print(f"cost = {cost}")
```

```python
def calculate_cost(a, b):
    n = len(df)
    print(f"n={n}")
    square_sum = 0 
    
    for i in range(n):
        x = df.loc[i, "temperature"]
        y = df.loc[i, "sales"]
        print(f"x={x}, y={y}")
        pred = a * x + b
        print(f"pred = {pred}")
        
        error = y - pred
        
        print(f"error = {error}")
        error_square = error ** 2
        print(f"error_square = {error_square}")
        square_sum  += error_square
        print("=" * 30)
    cost = square_sum / n
    return cost 
```

```python
cost1 = calculate_cost(-4, 80)
print(f"cost1 = {cost1}")
```

```python
cost2 = calculate_cost(-4.9, 81)
print(f"cost2 = {cost2}")
```

```python
a = -4 # 초기 값 (덜 팔리는 정도)
b = 80 # 초기 값 (바이어스)

learning_rate = 0.001

epochs = 1000   # 반복 횟수

n = 4          # 데이터 수 
cost_history = []

initial_a = a
initial_b = b

initial_cost = calculate_cost(a, b)
print(f"학습 전 cost = {initial_cost}")
```

```python
for epoch in range(epochs): 
    grad_a = 0 
    grad_b = 0
    
    for i in range(4):
        x = df.loc[i, "temperature"] 
        y = df.loc[i, "sales"]
        print(f"x={x}, y={y}")
        
        pred = a * x + b
    
        grad_a = grad_a + (pred - y) * x
        
        grad_b = grad_a + (pred - y)     #예측에서 찐을 뺀 것을 더해줌
    
    n = 4      
    grad_a = grad_a / n
    print(f"미분 결과 grad_a= {grad_a}")  #여기서 증가 시 a 에서 감소 / 여기서 감소 시 a 에서 증가
    
    grad_b = grad_b / n
    print(f"미분 결과 grad_b= {grad_b}")  #여기서 증가 시 a 에서 감소 / 여기서 감소 시 a 에서 증가
    
    a = a -1*(learning_rate * grad_a)
    print(f"update a = {a}")
    
    b = b - 1 * (learning_rate * grad_b)
    print(f"update b = {b}")
    
    cost = calculate_cost(a, b)
    cost_history.append(cost)
    
    print(f"cost = {cost}")
    print("=" * 30)
```
