```python
import pandas as pd

df = pd.DataFrame({
    "temperature": [10, 7, 4, 0],
    "sales": [30, 50, 60, 80]
})
display(df)
```

```python
X = df[["temperature"]].values.astype("float32")   #줄 번호 없앤다! : .values / 입력과 찐의 값이 정수여야 함: .astype("float32")

X           # 2 차원배열로 하고싶다! : [ ] 괄호 하나 더 씌우기
```

```python
y = df[["sales"]].values.astype("float32")   #줄 번호 없앤다! : .values / 입력과 찐의 값이 정수여야 함: .astype("float32")

y           # 2 차원배열로 하고싶다! : [ ] 괄호 하나 더 씌우기
```

```python
# X 값은 입력 / y 는 찐 값
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
X
```

```python
y
```

```python
model = keras.Sequential([
    layers.Dense(1, input_shape=(1,)) #교안 상에는 빠져있는 내용
])  #└ 선형회귀   /    └입력값 x 의 칸, 줄 수  → 줄은 생략 가능 → x → 1칸 4줄
```

```python
model.summary()
```

```python
model.compile(
    optimizer = keras.optimizers.SGD(learning_rate= 0.01),
    loss = "mse"
    )
```

```python
history = model.fit(X, y, epochs = 10000, verbose = 0) # 입력값 / 찐   어떻게 학습이 되는지 확인 : history 에 닮기
```

```python
history.history["loss"][-1]  #cost 가 어떻게 변하는지 확인  
```

```python
weight, bias = model.layers[0].get_weights()

a_value = weight[0][0]
b_value = bias[0]
print(f"weight={weight[0][0]}")
print(f"bias={bias[0]}")
```
