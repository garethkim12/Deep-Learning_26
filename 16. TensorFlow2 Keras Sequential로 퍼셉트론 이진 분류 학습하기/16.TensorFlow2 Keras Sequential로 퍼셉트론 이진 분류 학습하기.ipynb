```python
import numpy as np

import tensorflow as tf

import pandas as pd
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape = 1, ), # 입력: x_norm 칸수 (1칸)
    tf.keras.layers.Dense(1)
])
```

```python
model.summary()
```

```python
X = np.array([160, 170, 180, 190], dtype = np.float32).reshape(-1, 1)
print("입력값 X =", X)
```

```python
y = np.array([0, 0, 1, 1], dtype = np.float32).reshape(-1, 1)
print("정답값 y =", y)
```

```python
X_mean = X.mean()
```

```python
X_std = X.std()
```

```python
X_norm = (X - X_mean) / X_std 
```

```python
H_before = model(X_norm) #선형회귀 값
H_before
```

```python
z_before = tf.sigmoid(H_before)

z_before
```

```python
H_before.numpy() # 파이토치에선 값만 보려면 .item()을 썼으나, tensorflow 에선 .numpy()로 볼 수 있음. 추가로 넘파이 뒤에 [:]를 사용하면 필터링 가능
```

```python
lr = 0.1
ep = 1000

model.compile(  # H 는 선형회귀 값이지 확률이 아니다!
     optimizer = tf.keras.optimizers.SGD(learning_rate = lr)   # 한번에 기입!
    ,loss = tf.keras.losses.BinaryCrossentropy(from_logits= True)
) 
```

```python
history = model.fit( # 해당 반복 원할 시 .history 할 것.
    X_norm,
    y,
    epochs = ep,
    verbose = 1
)
```

```python
list(enumerate(history.history["loss"])) #enumerate 는 인덱스 , 전부 보려면 앞에 list 씌우면 됨.
```

```python
for epoch_index, mean_cost in enumerate(history.history["loss"]):
    if epoch_index % 100 == 0 or epoch_index == ep -1:
        print(f"epoch = {epoch_index}, mean_cost = {mean_cost}") # 에폭 인덱스에 대입 / mean_cost는 history loss 값
```
