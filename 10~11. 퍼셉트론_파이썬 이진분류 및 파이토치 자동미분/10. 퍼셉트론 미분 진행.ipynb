```python
import numpy as np
from math import e
```

```python
X = np.array([160, 170, 180, 190])

X
```

```python
y = np.array([0, 0, 1, 1])

y
```

```python
X_mean = np.mean(X) # 평균값
X_mean
```

```python
X_std = np.std(X) # 표준편차
X_std
```

```python
X_norm = (X-X_mean) / X_std  # 정규화 공식 (X 에서 평균 빼준 값을 표준편차로 나눈 값)
X_normㅋ
```

```python
a = 0.1
b = 0.0
```

```python
def sigmoid(H):
    return 1 / (1+ e**(-H))
```

```python
H_X = X * a + b

H_X
```

```python
Z_X = sigmoid(H_X)
Z_X
```

```python
a_미분값 = np.mean((Z_X - y ) * X) 

a_미분값
```

```python
epochs = 1000
epsilon = 1e-7 # 주의 : 붙여서 써야 인식함.
```

```python
epsilon
```

```python
1 - epsilon
```

```python
learning_rate = 0.1 # 학습 가중치
epochs = 1000
```

```python
for epoch in range(10):
    H = a * X_norm + b
    print(f"H = {H}") # 선형 회귀 값

    z = sigmoid(H)
    print(f"z = {z}") # 시그모이드 값
    
    grad_a  = np.mean((z - y) * X_norm)
    print(f"grad_a = {grad_a}")
    
    grad_b = np.mean((z-y))
    print(f"grad_b = {grad_b}")
    
    a = a - learning_rate * grad_a
    b = b - learning_rate * grad_b
    
    print(f"update a = {a}, update b = {b}")
    # z 가 epsilon 보다 작으면 epsilon 리턴 → - 무한대 epsilon 리턴 → 10 - 7
    # z 가 epsilon 보다 작으면 1 - epsilon 리턴 → 무한대 1 - epsilon 리턴 → 0.9999 리턴
    z_safe = np.clip(z, epsilon, 1- epsilon) # 0 값을 방지 하기 위함
    costs = -(y * np.log(z_safe) + (1-y)* np.log(1-z_safe))
    # costs = -(y * np.log(z) + (1-y)* np.log(1-z))
    mean_cost = np.mean(costs)
    print(f"mean_cost = {mean_cost}")
    print("="*30) 
```
