```python
#파이썬에서 수학적인 연산을 쉽게 하도록 제공하는 math 라이브러리에서 
# 로그(Logarithm)를 계산할 수 있는 log 함수를 가져옵니다.
from math import log
```

```python
# log(x, base) 함수 설명:
# 이 함수는 "base(밑)" 를  몇 번 곱해야 "x" 가 되는지 를 알려주는 로그 값을 계산합니다.
# 즉, log(base)를 기준으로 값(x)의 로그 값을 구하는 함수 입니다.
# 예를 들어, log(8,2)는 "2를 몇 번 곱하면 8이 되는지"를 계산합니다.
```

```python
# log(x, base) 함수의 매개변수 설명:
# - x: 로그 값을 구하려는 숫자 입니다.
#       이 숫자는 base(밑)를 몇번 곱해서 만들 수 있는지 궁금한 숫자 입니다.
# - base: 로그의 밑 (기준) 이 되는 숫자입니다.
#       이 숫자를 몇 번 곱해야 x 가 되는지 계산할 떄 기준으로 사용됩니다.

# 사용예시를 통해 자세히 살펴보겠습니다.
```

```python
# 예시
# 밑(base)이 2 인 로그에서 숫자 8의 로그 값을 계산합니다.
# 쉽게 말해서, 숫자 2를 몇 번 곱하면 8 이 되는지 구하는 것 입니다.
log_value = log(8,2)
```

```python
# 위의 게산 결과를 출력하여 확인합니다.
print(log_value)
# 출력 결과는 3.0 입니다.
# 이유는 다음과 같습니다:
# 2 x 2 x 2 = 8 → 즉, 숫자 2를 3번 곱하면 8이 되므로, log(8,2)의 값은 3입니다.
```

```python
# 예시 
# 이번에 밑(base)이 10인 로그에서 숫자 1000의 로그값을 계산합니다.
# 즉, 숫자 10을 몇 번 곱하면 1000이 되는지 계산하는 것입니다.
log_value2 = log(1000, 10)
```

```python
# 이 결과 역시 출력하여 확인합니다.
print(log_value2)
# 출력 결과는 3.0 입니다. 
# 이유는 다음과 같습니다.
# 10 x 10 x 10 = 1000 → 숫자 10을 3번 곱하면 1000이 되므로, log(1000, 10)의 값은 3입니다.
```

# 자연로그의 예시

```python
# math는 파이썬에서 수학 계산을 도와주는 라이브러리 입니다.
# 여기서 log 함수를 가져옵니다.
# log 함수는 로그 (logrithm) 값을 계산할 떄 사용합니다.
from math import log
```

```python
# log 함수에 대한 자세한 설명:
# log(x) 는 자연상수 e를 밑으로 하는 자연로그를 계산합니다.
# 매개변수 x : 로그값을 계산하고자 하는 숫자입니다.
```

```python
# 예시 1: 자연상수 e 의 자연로그 계산
# 자연상수 e의 값은 약 2.718281828459045 입니다.
# log(자연상수 e)는 "자연상수 e 를 몇번 곱해야 자연상수 e 가 되는가?" 라는 의미 입니다.
log_value = log(2.718281828459045) # log 함수로 e의 로그값을 계산합니다.
print(log_value)                   # 계산된 값을 출력하여 확인합니다.
# 출력 결과: 약 1.0 입니다.
# 이유: 자연상수 e를 정확히 1번 곱하면 e 가 되기 떄문입니다.
```

```python
# 예시 1: e³ 의 자연로그 계산
# e³(자연상수 e를 3번 곱한 값)의 자연로그 값을 계산합니다.
# "자연상수 e를 몇 번 곱하면 e³ 이 되는 가?" 라는 의미 입니다.
log_value = log(2.718281828459045**3) # e³ 값을 계산한 후 자연로그를 계산합니다.
print(log_value)                   # 결과값을 출력합니다.
# 출력 결과: 약 3.0 입니다.
# 이유:  e x e x e = e³ 이므로  자연상수 e를 3번 곱하면 e³ 입니다.
```

```python
# 예시 3: 숫자 1 의 자연로그 계산
# "자연상수 e를 몇 번 곱해야 숫자 1이 될까요?" 라는 질문입니다.
log_value = log(1) # 숫자 1의 자연로그 값을 계산합니다.
print(log_value)                   # 결과값을 출력합니다.
# 출력 결과: 약 0.0 입니다.
# 이유: 어떤 숫자던 0번 곱하면 결과는 항상 1이 되기 떄문입니다. (즉 e＾0 = 1)
```

```python
# 예시 3: 숫자 0의 자연로그 계산 (특별한 경우)
# "자연상수 e를 몇 번 곱해야 숫자 0이 될까요?" 라는 질문입니다.
# 수학적으로 자연상수 e를 곱하거나 나누어도 정확히 0이 될 수 없습니다.
# 이 경우는 수학적으로 음의 무한대 이며, 파이썬 코드에서 오류가 발생합니다.
log_value = log(0) # 숫자 0의 자연로그 값을 계산합니다.
print(log_value)   # 이 코드는 오류가 발생합니다. 
# 출력 결과: 오류(valueError)가 발생합니다.
# 이유: 음의 무한대는 파이썬에서 정의되지 않기 떄문입니다.
```

# 퍼셉트론

```python
# numpy는 숫자 계산을 빠르고 쉽게 도와주는 파이썬 라이브러리 입니다.
# 인공지능 코드에서는  배열, 행렬, 수학, 함수 계산이 자주 필요하기 때문에 numpy를 많이 사용합니다.
# as np는 numpy를 앞으로 np 라는 짧은 이름으로 사용하겠다는 뜻입니다.
import numpy as np
from math import e
```

```python
# X는 입력값입니다.
# 여기서는 사람의 키(cm)를 입력값으로 사용합니다.
# 머신러닝에서는 사람이 정리한 표(csv등) 의 각 컬럼 값이 모델의 입력값으로 사용되는 경우가 많습니다.
# np.array()는 여러 개의 숫자를 하나의 배열로 묶어주는 함수 입니다.
X = np.array([160, 170, 180, 190])

X
```

```python
# y는 실제 정답값입니다.
# 이 예제에서는 사람이 농구선수인지 아닌지를 0 또는 1로 표시합니다.
# 0은 농구선수가 아님, 1은 농구선수가 맞음을 의미합니다.
y = np.array([0, 0, 1, 1])

y
```

```python
# a는 가중치(weight)입니다.
# 가중치는 입력값 X 가 결과에 얼마나 영향을 주는지를 조절하는 값입니다.
a = 0.1

# b는 편향(bias)입니다.
# 편향은 전체 계산 결과의 기준선을 조정하는 값입니다.
b = -17
```

```python
# H = a * X + b
#
# 수식에서는 H(x) = ax + b 라고 씁니다.
# 하지만 코드에서는 변수 이름에 괄호를 사용할 수 없으므로 H 라고 저장합니다.
#
# H(x)는 선형 회귀에서 배운 ax + b 형태와 같은 선형 계산 결과 입니다.
# 단, 여기서는 회귀 문제처럼 H(x)를 최종 예측값으로 사용하지 않습니다.
# 이진 분류에서는 H(x)를 시그모이드 함수에 넣기 전의 "선형 점수"로 사용합니다.
H = a * X + b

print("선형 계산 결과 H(x):", H)
```

```python
# sigmoid() 함수는 H(x)를 0과 1 사이의 값으로 변환하는 함수 입니다.
# H(x)는 선형 계산 결과이므로 그대로는 확률로 해석하기 어렵습니다.
# 그래서 시그모이드 함수를 사용해 H(x)를 예측 확률값으로 바꿉니다.
def sigmoid(H):
    return 1 / (1 + e**(-H))
```

```python
# z 는 시그모이드 함수를 통과한 예측 확률값입니다.
#
# 수식으로 쓰면 다음과 같습니다.
# 
# z = sigmoid(H(x))
#
# z 는 항상 0과 1 사이의 값을 가집니다.
# z 가 1에 가까우면 모델은 y=1 일 확률이 높다고 예측한 것입니다.
# z 가 0에 가까우면 모델은 y=0 일 확률이 높다고 예측한 것입니다.
z = sigmoid(H)

print("시그모이드 예측확률: z:", z)
```

```python
# Binary Cross Entropy (BCE)는 이진 분류에서 자주사용하는 비용함수입니다.
# BCE 는 예측 확률값 z 와 실제 정답 y 의 차이를 비용(Cost)으로 계산합니다.
# 
# BCE 공식:
#
# C(z, y) = -y * log(z) - (1 - y) * log(1 - z)
#
# y가 1이면 첫번째 항 -y * log(z) 가 작동합니다.
# 이때 z가 1에 가까울수록 -log(z)는 0에 가까워져 비용이 작아집니다.
#
# y가 0이면 두 번째 항 -(1 - y) * log(1 - z)가 작동합니다.
# 이때 z가 0에 가까울수록 -log(1 - z)는 0에 가까워져 비용이 작아집니다.
costs = - (y  * np.log(z) + (1 - y) * np.log(1 - z))

print("각 샘플의 비용(Cost):", costs)
```

```python
# 평균 비용을 계산합니다.
# costs에는 각 샘플별 비용이 들어 있습니다.
# 모델 전체의 성능을 하나의 숫자로 보고 싶다면 각 샘플의 비용을 평균 냅니다.
mean_cost = np.mean(costs)

# 이 교재에서는 Cost라는 용어로 통일합니다.
print("평균 비용(Cost):", mean_cost)
```
